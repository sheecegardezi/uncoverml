```python
from pathlib import Path
from pprint import pprint
```

```python
input_file_path = Path('../data/feature_elimination_results.txt')
```

```python
results = {}
```

```python
myfile = open(input_file_path, "r")
lines = myfile.readlines()

```

```python
results = {}
for line in lines:
    line  = line.strip()
    if "Staring experiment for:" in line:
        experiment_id = int((line.split("Staring experiment for:  ")[1]))
        results[experiment_id] = {}
        results[experiment_id]["scores"]={}
    elif "Getting score for:" in line:
        _,_,feature_name,feature_score,_=line.split("Getting score for:")[1].split(" ")
        results[experiment_id]["scores"][feature_name] = round(float(feature_score),4)
    elif "Results of  " in line:
        _,_,_,_,_,_, lowest_score, lowest_feature = line.split(" ")
        results[experiment_id]["lowest_score"] = round(float(lowest_score),4)
        results[experiment_id]["lowest_feature"] = lowest_feature
    elif "Wall time: " in line:
        print(line)
```

```python
myfile.close()
```

```python
x = []
y = []
ymax=[]
ymin=[]
yerr =[]
for experiment_id in results.keys():
    
    scores = []
    for feature_name in results[experiment_id]["scores"].keys():
        scores.append(results[experiment_id]["scores"][feature_name])
        
    x.append(experiment_id)
    y.append(results[experiment_id]["lowest_score"])
    yerr1 = max(scores) - min(scores)
    yerr.append(yerr1)
    
x,y,yerr
```

```python
 %matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = "18"
plt.rcParams['figure.dpi'] = 200 
plt.rcParams['figure.figsize'] = [20, 15]
plt.suptitle('Feature Elimination', fontsize=20)
plt.xlabel('Number of features model trained with', fontsize=20)
plt.ylabel('Model performance score (rmsle)', fontsize=20)
plt.style.use('seaborn-whitegrid')
plt.errorbar(x, y, yerr=yerr, fmt='.k');
plt.xlim(50,0) 
# plt.ylim(0) 

plt.show()
```
