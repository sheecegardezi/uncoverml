```python
from sklearn import metrics


scorings = ['explained_variance', 'max_error', 'neg_mean_absolute_error', 'neg_mean_squared_error', 'neg_root_mean_squared_error', 'neg_mean_squared_log_error', 'neg_median_absolute_error', 'r2', 'neg_mean_poisson_deviance', 'neg_mean_gamma_deviance', 'neg_mean_absolute_percentage_error']
y_test, y_test_max,y_test_min = [1,2,3,4,5,6],[1,2,3,4,5,6],[900, 800,100,222,331,143]

for scoring in scorings:
	if scoring == 'explained_variance':
		score_max = metrics.explained_variance_score(y_test, y_test_max)
		score_min = metrics.explained_variance_score(y_test, y_test_min)
	if scoring == 'max_error':
		score_max = metrics.max_error(y_test, y_test_max)
		score_min = metrics.max_error(y_test, y_test_min)
	if scoring == 'neg_mean_absolute_error':
		score_max = metrics.mean_absolute_error(y_test, y_test_max)
		score_min = metrics.mean_absolute_error(y_test, y_test_min)
	if scoring == 'neg_mean_squared_error':
		score_max = metrics.mean_squared_error(y_test, y_test_max)
		score_min = metrics.mean_squared_error(y_test, y_test_min)
	if scoring == 'neg_root_mean_squared_error':
		score_max = metrics.mean_squared_error(y_test, y_test_max)
		score_min = metrics.mean_squared_error(y_test, y_test_min)
	if scoring == 'neg_mean_squared_log_error':
		score_max = metrics.mean_squared_log_error(y_test, y_test_max)
		score_min = metrics.mean_squared_log_error(y_test, y_test_min)
	if scoring == 'neg_median_absolute_error':
		score_max = metrics.median_absolute_error(y_test, y_test_max)
		score_min = metrics.median_absolute_error(y_test, y_test_min)
	if scoring == 'r2':
		score_max = metrics.r2_score(y_test, y_test_max)
		score_min = metrics.r2_score(y_test, y_test_min)
	if scoring == 'neg_mean_poisson_deviance':
		score_max = metrics.mean_poisson_deviance(y_test, y_test_max)
		score_min = metrics.mean_poisson_deviance(y_test, y_test_min)
	if scoring == 'neg_mean_gamma_deviance':
		score_max = metrics.mean_gamma_deviance(y_test, y_test_max)
		score_min = metrics.mean_gamma_deviance(y_test, y_test_min)
	if scoring == 'neg_mean_absolute_percentage_error':
		score_max = metrics.mean_absolute_percentage_error(y_test, y_test_max)
		score_min = metrics.mean_absolute_percentage_error(y_test, y_test_min)
	print(scoring, score_max, score_min, score_max >= score_min)
```
